```python
import os
import matplotlib.pyplot as plt
```

```python
def count_images_in_subfolders(folder_path):
    """
    Count the number of images in each subfolder of a given folder.

    Parameters:
        folder_path (str): Path to the folder containing subfolders.

    Returns:
        dict: A dictionary where keys are subfolder names and values are the number of images in each subfolder.
    """
    image_counts = {}
    
    # Iterate through each subfolder in the main folder
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        
        # Check if the subfolder is a directory
        if os.path.isdir(subfolder_path):
            # Count the number of image files in the subfolder
            image_count = sum(1 for file in os.listdir(subfolder_path) if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')))
            image_counts[subfolder] = image_count
    
    return image_counts
```

```python
def plot_bar_graph(image_counts, folder_name):
    """
    Plot a bar graph of the number of images in each subfolder.

    Parameters:
        image_counts (dict): A dictionary where keys are subfolder names and values are the number of images in each subfolder.
        folder_name (str): Name of the folder being plotted.
    """
    subfolders = list(image_counts.keys())
    counts = list(image_counts.values())

    # Define colors for each subfolder
    colors = ['#491D8B', '#6929C4', '#8A3FFC', '#A56EFF']

    plt.figure(figsize=(12, 6))
    bars = plt.bar(subfolders, counts, color=colors)
    plt.xlabel('Subfolder')
    plt.ylabel('Number of Images')
    plt.title(f'Number of Images in {folder_name}')
    plt.xticks(rotation=45, ha='right')

    # Display image counts on the bars
    for i, count in enumerate(counts):
        plt.text(i, count + 5, str(count), ha='center', va='bottom', fontsize=8, color='black')

    plt.tight_layout()
    plt.show()
```

```python
# Test folder
folder_path = r'D:\Git Projects\Peppermint-Identification-and-Classification-Model\Identification\Data\test'
image_counts = count_images_in_subfolders(folder_path)
print("Number of images in each subfolder:", image_counts)
folder_name = 'Test'
plot_bar_graph(image_counts,folder_name)
```
